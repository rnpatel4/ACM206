```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy
```

## MCMC For a Gaussian

Design a M-H sampler for the N(0,1) distribution

```python
x = 0    #initial value
x_arr = []
s = 0.9

for i in range(10000):
    proposal = x + s*np.random.uniform(-1,1)
    k = proposal
    U = np.random.uniform()
    # if k < 0:
    #     continue
    # try:
    prob_proposal = scipy.stats.norm(0, 1).pdf(proposal)
    prob_existing = scipy.stats.norm(0, 1).pdf(x)
    if (1/2)*(prob_proposal/prob_existing >= U):
        x = proposal
    else:
        x = x
    # except:
    #     x = x
    # finally:
    x_arr = np.append(x_arr, x)
```

```python
plt.plot(x_arr)
```

```python
np.mean(x_arr)
```

```python
np.var(x_arr)
```

```python
plt.hist(x_arr)
```

## MCMC for two Gaussians

Design M-H Sampler for a mixture of two Gaussians: 1/2 N(-1,1) + 1/2 N(1,1)

```python
x = 0    #initial value
x_arr = []
s = -0.75

for i in range(100000):
    proposal = x + s*np.random.uniform(-1,1)
    k = proposal
    U = np.random.uniform()
    # if k < 0:
    #     continue
    # try:
    prob_proposal = 1/2*(scipy.stats.norm(-1, 1).pdf(proposal) + scipy.stats.norm(1, 1).pdf(proposal))
    prob_existing = 1/2*(scipy.stats.norm(-1, 1).pdf(x) + scipy.stats.norm(1, 1).pdf(x))
    if (1/2)*(prob_proposal/prob_existing >= U):
        x = proposal
    else:
        x = x
    # except:
    #     x = x
    # finally:
    x_arr = np.append(x_arr, x)
```

```python
plt.plot(x_arr)
```

```python
binwidth = 0.2
bins=np.arange(min(x_arr), max(x_arr) + binwidth, binwidth)
plt.hist(x_arr, bins=bins)
```

```python
np.mean(x_arr)
```
