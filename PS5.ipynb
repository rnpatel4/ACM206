## Poisson Sampling

Design a Metropolis sampler that targets the Pois(λ) distribution on the nonnegative integers {0, 1, 2, . . .}, which has probability mass function p(x) = e^−λ * λ^k/k!. Implement your sampler with λ = 3, 5, 7. How well does it work?

```python
lbda = 100 #lambda
x = 100    #initial value
x_arr = []
for i in range(100000):
    proposal = int(np.random.normal(loc=x, scale=10))
    k = proposal
    U = np.random.uniform()
    # if k < 0:
    #     continue
    try:
        prob_proposal = lbda**k * np.exp(-lbda)/np.math.factorial(k)
        prob_existing = lbda**x * np.exp(-lbda)/np.math.factorial(x)
        if (1/2)*(prob_proposal/prob_existing >= U):
            x = proposal
        else:
            x = x
    except:
        x = x
    finally:
        x_arr = np.append(x_arr, x)
```
